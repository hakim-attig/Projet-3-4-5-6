## API OpenFoodFacts 

```python
import requests
import pandas as pd

# === URL de l'API OpenFoodFacts ===
url = "https://world.openfoodfacts.org/cgi/search.pl"

# === Paramètres de requête ===
params = {
    'search_terms': 'champagne',
    'search_simple': 1,
    'action': 'process',
    'json': 1,
    'page_size': 50  # Tu peux augmenter si besoin
}

# === Requête GET ===
response = requests.get(url, params=params)

# === Vérification de la réponse ===
if response.status_code == 200:
    print("Requête réussie")
    data = response.json()
    products = data.get('products', [])
else:
    print(f"Erreur {response.status_code}")
    print(response.text)
    products = []

# === Extraction des données ===
filtered_data = []

for prod in products:
    food_id = prod.get('id', '')
    label = prod.get('product_name', '')
    category = prod.get('categories', '')
    food_contents = prod.get('ingredients_text', '')
    image_url = prod.get('image_url', '')

    # On filtre pour s'assurer que "champagne" apparaît bien dans les ingrédients
    if food_contents and 'champagne' in food_contents.lower():
        filtered_data.append({
            'foodId': food_id,
            'label': label,
            'category': category,
            'foodContentsLabel': food_contents,
            'image': image_url
        })

# === Création du DataFrame ===
df_openfood = pd.DataFrame(filtered_data)
print(df_openfood.head())

# === Sauvegarde CSV ===
df_openfood.to_csv("produits_champagne_openfoodfacts.csv", index=False)
print("Fichier CSV sauvegardé : produits_champagne_openfoodfacts.csv")
```
